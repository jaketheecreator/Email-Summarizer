```python
import re
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.luhn import LuhnSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
from textblob import TextBlob
import nltk

# This is the function to summarize the text
def summarize_text(text, language="english", sentence_count=3):
    """
    Summarizes a given text using the Luhn summarization algorithm.
    """
    parser = PlaintextParser.from_string(text, Tokenizer(language))
    stemmer = Stemmer(language)
    summarizer = LuhnSummarizer(stemmer)
    summarizer.stop_words = get_stop_words(language)

    summary = summarizer(parser.document, sentence_count)
    return " ".join([str(s) for s in summary])

# This is the function to generate a reply based on the summary
def generate_smarter_reply(text):
    """
    Generates a reply based on the sentiment of the original text.
    """
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    
    if sentiment_score > 0.1:
        return """
Hi,

Thanks for the positive update! It sounds like things are going well.

Best,
[Your Name]
"""
    elif sentiment_score < -0.1:
        return """
Hi,

Thanks for the update. I'm sorry to hear about the challenges. Let's discuss this further.

Best,
[Your Name]
"""
    else:
        return """
Hi,

Thanks for the update. It sounds like things are on track.

Best,
[Your Name]
"""

# This is the NEW function to extract calendar information
def extract_calendar_info(text):
    """
    Extracts potential meeting dates and times from the text.
    Note: This is a simple implementation and may not catch all patterns.
    """
    date_patterns = [
        r"(next\s+(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday))",
        r"(by\s+\w+\s+\d{1,2})",
        r"(scheduled\s+for\s+\w+\s+\d{1,2})"
    ]
    time_patterns = [
        r"(at\s+\d{1,2}:\d{2}(?:\s*am|\s*pm)?)",
        r"(by\s+\d{1,2}:\d{2}(?:\s*am|\s*pm)?)",
        r"(EOD\s+Friday)",
        r"(by\s+Tuesday\s+morning)"
    ]

    dates_found = []
    for pattern in date_patterns:
        dates_found.extend(re.findall(pattern, text, re.IGNORECASE))

    times_found = []
    for pattern in time_patterns:
        times_found.extend(re.findall(pattern, text, re.IGNORECASE))
        
    return dates_found, times_found

# --- Start of the user-facing part ---

# 1. Define the email or long text you want to summarize
my_long_email = """
Subject: Project Alpha - Weekly Update & Action Items

Hi Team,

I hope you're all having a productive week. I'm writing to provide a detailed update on the status of Project Alpha and to outline the next steps.

Our team has made significant progress over the past week. The initial wireframes for the user dashboard have been completed and approved by the stakeholders. The design team is now moving on to creating the high-fidelity mockups.

On the development front, the back-end team has successfully implemented the user authentication module and is currently working on the API endpoints for data retrieval. The front-end team has started building the core components based on the initial wireframes. We encountered a minor bug with the user login functionality, which John has already been assigned to fix.

The marketing and content teams have begun drafting the copy for the product landing page and have scheduled a meeting for next Tuesday to finalize the messaging. I've attached the draft documents for your review.

Action Items:
1. All team members, please review the attached content draft by EOD Friday.
2. The front-end team should sync with the back-end team by Wednesday to ensure a smooth integration process.
3. John, please provide an update on the login bug fix by Tuesday morning.
4. The design team should have the high-fidelity mockups ready for review by the end of next week.

Overall, we are on track with our timelines. Great work, everyone.

Best,
[Your Name]
"""

# 2. Call all the functions
summary = summarize_text(my_long_email, sentence_count=2)
smarter_reply = generate_smarter_reply(my_long_email)
dates, times = extract_calendar_info(my_long_email)

# 3. Print the results
print("Original Text:")
print(my_long_email)
print("\nSummary:")
print(summary)
print("\nGenerated Smart Reply:")
print(smarter_reply)
print("\n--- Potential Calendar Events ---")
print(f"Dates Found: {', '.join(dates) or 'None'}")
print(f"Times Found: {', '.join(times) or 'None'}")
```
